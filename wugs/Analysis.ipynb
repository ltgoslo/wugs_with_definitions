```python
import pandas as pd
from pathlib import Path
import csv
```

```python
dfs = []
for p in Path('./').glob('rushifteval_public/*/*/data/*/uses.csv'):
    df = pd.read_csv(p, quoting=csv.QUOTE_MINIMAL, sep='\t')
    ds = p.parent.parent.parent.name
    df['part'] = ds
    df['lang'] = 'ru'
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df.drop(df.index[df.indexes_target_token=='0:0'], inplace=True) # TODO: fix in the dataset!
df['positions'] = df['indexes_target_token'].str.split(':').apply(lambda l: list(map(int, l)))
df.head()
```

```python
dfs = []
for p in Path('./').glob('nor_dia_change/*/data/*/uses.csv'):
    df = pd.read_csv(p, quoting=csv.QUOTE_MINIMAL, sep='\t')
    ds = p.parent.parent.parent.name
    df['part'] = ds.replace('subset','nordiachange')
    df['lang'] = 'no'
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df.drop(df.index[df.indexes_target_token=='0:0'], inplace=True)  # TODO: fix in the dataset!
df['positions'] = df['indexes_target_token'].str.split(':').apply(lambda l: list(map(int, l)))
df['positions'] = df['positions'].apply(lambda l: [l[0], l[1]+1])  # TODO: fix in the dataset!
df.head(3)
```

```python
dfs = []
for p in Path('./').glob('dwug_*/data/*/uses.csv'):
    df = pd.read_csv(p, quoting=csv.QUOTE_NONE, sep='\t')
    ds = p.parent.parent.parent.name
    df['part'] = ds
    df['lang'] = ds.split('_')[1]
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df['positions'] = df['indexes_target_token'].str.split(':').apply(lambda l: map(int, l))
df.head(3)
```

### sanity check the specified positions of the target words

```python
df['wordat'] = df.apply(lambda r: r.context[slice(*r.positions)], axis=1)
df['wordat_ctxed'] = df.apply(lambda r: r.context[max(0,r.positions[0]-5):r.positions[1]+5 ], axis=1)
forms = df.groupby('lemma').wordat.apply(lambda s: 
                                        ', '.join(f'{k} {v}' for k,v in s.value_counts().to_dict().items())).reset_index()
pd.set_option('colwidth',300)
pd.set_option('display.max_rows',300)
forms
```

```python
# !pip install simplemma
from simplemma import lemmatize
```

```python
q = df.groupby(['lemma','lang']).apply(lambda g: ' '.join(set(
    lemmatize(f, lang='nb' if g.lang.iloc[0]=='no' else g.lang.iloc[0], greedy=True) 
    for f in g.wordat.unique()
))).reset_index()
```

```python
m0 = (q[0].str.split().str.len() == 1) & (q[0].str.split().str[0]==q.lemma.str.split('_').str[0])
print('1 normal form which is equal to lemma:', m0.mean())
m1 = q[0].str.split().str.len() > 1
print('More than 1 normal form found:',m1.mean())
m2 = (q[0].str.split().str.len() == 1) & (q[0].str.split().str[0]!=q.lemma.str.split('_').str[0])
print('1 normal form which is not lemma', m2.mean())
```
